Cell 1: What is a Variable?

```python
# A variable is like a labeled box where we store data

name = "Vikram"          # Text (string)
age = 25                 # Whole number (integer)
height = 5.9             # Decimal number (float)
is_student = True        # True or False (boolean)

print(name)
print(age)
print(height)
print(is_student==0)
print(is_student==1)
```

Cell 2: Variable Naming Rules (Valid vs Invalid)

```python
# ✅ VALID variable names
student_name = "Aarav"
_roll_no = 101
marks_2026 = 85
totalScore = 450

print(student_name, _roll_no, marks_2026, totalScore)

# ❌ INVALID variable names (will give SyntaxError)
#1name = "Ravi"          # Cannot start with number
#student-name = "Priya"  # Hyphen not allowed
#class = "CSE"           # 'class' is a reserved keyword
#my name = "Vikram"      # Space not allowed
```

Cell 3: Dynamic Typing in Python

```python
# Python variables can change their type anytime (Dynamic Typing)

x = 10
print("Value:", x, "| Type:", type(x))

x = 10.5
print("Value:", x, "| Type:", type(x))

x = "Hello"
print("Value:", x, "| Type:", type(x))

x = True
print("Value:", x, "| Type:", type(x))
```

Cell 4: Using type() Function

```python
a = 100
b = 99.5
c = "ACE College"
d = False
e = [1, 2, 3]          # We will study lists later

print("a is of type:", type(a))
print("b is of type:", type(b))
print("c is of type:", type(c))
print("d is of type:", type(d))
print("e is of type:", type(e))
```

Cell 5: f-strings (Recommended Way)

```python
name = "Shireesh"
age = 23

percentage = 87.4567

# Basic f-string
print(f"Student Name: {name}")
print(f"Age: {age}")

# f-string with expressions
print(f"Next year {name} will be {age + 1} years old")

# Formatting numbers
print(f"Percentage: {percentage:.2f}%")        # 2 decimal places
print(f"Percentage: {percentage:.0f}%")        # No decimal places
print(f"Percentage: {percentage:.1f}%") 
# Adding text and alignment
print(f"Name: {name:<10} | Age: {age:>3}")     # Left & right alignment
```

Cell 6: Combining All Concepts (Mini Practical Example)

```python
# Student details using variables + f-strings + type checking

student_name = "Rahul Verma"
roll_number = 103
current_semester = 4
cgpa = 8.75
is_placed = False

print("=== Student Information ===")
print(f"Name           : {student_name}")
print(f"Roll No        : {roll_number}")
print(f"Semester       : {current_semester}")
print(f"CGPA           : {cgpa}")
print(f"Placement Status: {'Placed' if is_placed else 'Not Placed'}")

print("\n--- Data Types ---")
print(f"student_name is : {type(student_name)}")
print(f"roll_number is  : {type(roll_number)}")
print(f"cgpa is         : {type(cgpa)}")
print(f"is_placed is    : {type(is_placed)}")
```

Cell 8: Common Mistakes to Avoid

```python
print("=== Common Mistakes with Variables ===\n")

# Mistake 1: Using reserved keywords
# class = "CSE"          # ❌ SyntaxError: invalid syntax
student_class = "CSE"     # ✅ Correct
print("Mistake 1 fixed:", student_class)

# Mistake 2: Variable name starting with number
# 2nd_sem = "Even"       # ❌ SyntaxError
second_sem = "Even"       # ✅ Correct
print("Mistake 2 fixed:", second_sem)

# Mistake 3: Spaces or special characters in name
# my name = "Vikram"     # ❌ SyntaxError
my_name = "Vikram"        # ✅ Correct
print("Mistake 3 fixed:", my_name)

# Mistake 4: Case sensitivity
age = 21
# print(Age)             # ❌ NameError: name 'Age' is not defined
print("Mistake 4 fixed - age:", age)

# Mistake 5: Accidentally overwriting built-in functions
# type = "Student"       # ❌ Bad practice (shadows built-in type())
my_type = "Student"       # ✅ Correct
print("Mistake 5 fixed:", my_type)

# Mistake 6: Forgetting 'f' in f-string
name = "Priya"
# print("Hello {name}")  # ❌ Prints literally: Hello {name}
print(f"Hello {name}")    # ✅ Correct

# Mistake 7: Dynamic typing confusion
marks = 85
print("Before reassigning, type of marks:", type(marks))

marks = "Eighty Five"     # Now it becomes string
print("After reassigning, type of marks:", type(marks))
print("Value is now:", marks)

print("\n=== All common mistakes covered ===")
```
